```python
from pymongo import MongoClient

client = MongoClient("mongodb://db:27017/")
db = client.chelle

asset = db.raw_assets.find_one()
print(asset)
```

```python
from pprint import pprint

# Get and pretty print a sample document
first_doc = db.raw_assets.find_one({})
print("\nSample document structure:")
pprint(first_doc, indent=2, width=80)

# To see all unique fields across all documents (in case some docs have different fields)
all_fields = set()
for doc in db.raw_assets.find({}):
    all_fields.update(doc.keys())

print("\nAll unique fields in collection:")
pprint(sorted(list(all_fields)), indent=2)
```
